```python
import pandas as pd
df = pd.read_csv("/home/omkar/Omkar Pawar/Data Science/Placement Assignment Omkar Pawar/Statistics/data.csv")

df.head(3)
```

```python
import scipy.stats as stats
import seaborn as sns
import matplotlib.pyplot as plt

# Shapiro-Wilk Test
_, p_value = stats.shapiro(df[' Blood Pressure After (mmHg)'])
alpha = 0.05  # Significance level

if p_value > alpha:
    print("The column follows a normal distribution (p-value > 0.05)")
else:
    print("The column does not follow a normal distribution (p-value <= 0.05)")

# Histogram
sns.histplot(df[' Blood Pressure After (mmHg)'], kde=True)
plt.xlabel(' Blood Pressure After (mmHg)')
plt.ylabel("Frequency")
plt.title("Histogram of " + ' Blood Pressure After (mmHg)')
plt.show()

# Q-Q Plot
stats.probplot(df[' Blood Pressure After (mmHg)'], dist="norm", plot=plt)
plt.xlabel("Theoretical Quantiles")
plt.ylabel("Sample Quantiles")
plt.title("Q-Q Plot of " + ' Blood Pressure After (mmHg)')
plt.show()
```
